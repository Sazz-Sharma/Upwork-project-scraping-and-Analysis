```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

```python
df = pd.read_csv('job_data.csv')
```

```python


# skills = []
# for string_data in df['Skill Set'].values:
#     data_list = [item.strip() for item in string_data[1:-1].split(',')] 
#     for skill in data_list:
#         skills.append(skill)

skills = list(filter(lambda x: (x != "''" and not(x.startswith("'+"))),[skill.strip() for string_data in df['Skill Set'].values for skill in string_data[1:-1].split(',')]))
skill_counts = pd.Series(skills).value_counts()

# skill_counts.plot(kind='bar')
total_count = sum(skill_counts.values)
filtered_skills = {skill: count for skill, count in skill_counts.items() if count / total_count > 0.03}
other_count = total_count - sum(filtered_skills.values()
filtered_skills['Other'] = other_count
plt.figure(figsize=(8, 8))
plt.pie(filtered_skills.values(), labels=filtered_skills.keys(), autopct='%1.1f%%', startangle=140)
plt.show()
```
